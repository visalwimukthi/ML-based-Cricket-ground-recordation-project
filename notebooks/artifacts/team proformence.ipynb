```python
def recommend_grounds(user_features, date, team_name=None, city=None, top_n=5):

    data = df.copy()

    #  Filter by city
    if city:
        data = data[data["city"].str.lower() == city.lower()]

    #  Team performance logic
    if team_name:
        team_data = team_stats[team_stats["team_name"] == team_name].copy()

        team_data["win_rate"] = team_data["wins"] / team_data["matches_played"]

        data = data.merge(team_data, on="ground_name", how="left")

        data["win_rate"] = data["win_rate"].fillna(0).astype(float)
        data["matches_played"] = data["matches_played"].fillna(0).astype(float)

        data = data.drop_duplicates(subset=["ground_name"])

    else:
        data["win_rate"] = 0

    #  weather column
    data["weather"] = data["city"].apply(lambda x: get_weather(x, date))

    #  Weather scoring function 
    def weather_score(condition):
        condition = str(condition).lower()
        if "rain" in condition or "storm" in condition:
            return 0
        elif "cloud" in condition:
            return 0.7
        else:
            return 1

    data["weather_score"] = data["weather"].apply(weather_score)

    # Remove booked grounds
    if not bookings.empty:
        booked = bookings[bookings["date"] == date]["ground_name"]
        data = data[~data["ground_name"].isin(booked)]


    # ML SCORING
    

    categorical_cols = ["city", "district", "province", "ground_type", "pitch"]

    # Encode full dataset
    data_encoded = pd.get_dummies(data, columns=categorical_cols)

    # Align with trained model features
    data_encoded = data_encoded.reindex(columns=model_features, fill_value=0)

    # Convert to numeric
    data_encoded = data_encoded.apply(pd.to_numeric, errors='coerce').fillna(0)

    # Scale
    X_scaled = scaler.transform(data_encoded)

    # Predict
    scores = model.predict_proba(X_scaled)[:, 1]

    # Assign ML score
    data["ml_score"] = scores

    # Final hybrid score (WITH WEATHER)
    data["final_score"] = (
        data["ml_score"] * 0.4 +
        data["win_rate"] * 0.4 +
        data["weather_score"] * 0.2
    )

    # Return top results
    return data.sort_values(by="final_score", ascending=False).head(top_n)
```

```python
# Step 1: define input
user_input = {
    "price_(lkr)": 120000,
    "lights": 1,
    "parking": 1,
    "rating": 4
}

# Step 2: call function
result = recommend_grounds(
    user_input,
    date="2026-04-10",
    team_name="Royal College XI",
    city="colombo"
)

# step 3: output
top = result.iloc[0]
alt = result.iloc[1]

print("Ground Recommendation Result\n")

print("Best Match:")
print(f"{top['ground_name']} | Weather: {top['weather']}")

if "rain" in str(top["weather"]).lower():
    print("\n Not suitable due to bad weather")

    print("\n Alternative:")
    print(f"{alt['ground_name']} | Weather: {alt['weather']}")
else:
    print("\n Suitable for play")
```
