```python
import pandas as pd
import joblib

# Load dataset
df = pd.read_csv("../artifacts/Ground_clean_dataset.csv")

# Load trained ML components
rf_model = joblib.load("../artifacts/rf_model.joblib")
scaler = joblib.load("../artifacts/scaler.joblib")
model_features = joblib.load("../artifacts/model_features1.joblib")

print("Dataset and model loaded successfully")
```

```python
# Columns used in training
yes_no_cols = ["Lights", "Parking", "Washrooms", "Pavilion", "Scoreboard"]
numeric_cols = ["Length", "Width", "Rooms", "Parking Cap.", "Price (LKR)", "Rating"]
categorical_cols = ["City", "District", "Province", "Ground Type", "Pitch"]

# Pre-calc medians & modes
num_medians = {col: df[col].median() for col in numeric_cols}
cat_modes = {col: df[col].mode()[0] for col in categorical_cols}

print("Helper stats ready")
```

### normal Ground Search

```python
def search_grounds(
    city=None,
    district=None,
    province=None,
    ground_type=None,
    pitch=None,
    needs_lights=None,
    needs_parking=None,
    max_price=None
):

    results = df.copy()

    # Normalize text inputs to lowercase 
    city_l        = city.lower().strip()        if city else None
    district_l    = district.lower().strip()    if district else None
    province_l    = province.lower().strip()    if province else None
    ground_type_l = ground_type.lower().strip() if ground_type else None
    pitch_l       = pitch.lower().strip()       if pitch else None

    if city_l:
        results = results[results["City"].str.lower() == city_l]

    if district_l:
        results = results[results["District"].str.lower() == district_l]

    if province_l:
        results = results[results["Province"].str.lower() == province_l]

    if ground_type_l:
        results = results[results["Ground Type"].str.lower() == ground_type_l]

    if pitch_l:
        results = results[results["Pitch"].str.lower() == pitch_l]

    if needs_lights is not None:
        results = results[results["Lights"] == ("Yes" if needs_lights else "No")]

    if needs_parking is not None:
        results = results[results["Parking"] == ("Yes" if needs_parking else "No")]

    if max_price is not None:
        results = results[results["Price (LKR)"] <= max_price]

    if results.empty:
        print("No grounds found for these filters.")
        return

    cols_to_show = [
        "Ground Name",
        "City",
        "District",
        "Province",
        "Ground Type",
        "Pitch",
        "Lights",
        "Parking",
        "Price (LKR)",
        "Rooms",
        "Rating"
    ]
    return results[cols_to_show]
```

### ML-based Recommendation

```python
def recommend_grounds(
    city=None,
    district=None,
    province=None,
    ground_type=None,
    pitch=None,
    needs_lights=None,
    needs_parking=None,
    max_price=None,
    top_n=5
):
    subset = df.copy()

    # Normalize text inputs
    city_l        = city.lower().strip()        if city else None
    district_l    = district.lower().strip()    if district else None
    province_l    = province.lower().strip()   if province else None
    ground_type_l = ground_type.lower().strip() if ground_type else None
    pitch_l       = pitch.lower().strip()       if pitch else None

    # Apply filters
    if city_l:
        subset = subset[subset["City"].str.lower() == city_l]

    if district_l:
        subset = subset[subset["District"].str.lower() == district_l]

    if province_l:
        subset = subset[subset["Province"].str.lower() == province_l]

    if ground_type_l:
        subset = subset[subset["Ground Type"].str.lower() == ground_type_l]

    if pitch_l:
        subset = subset[subset["Pitch"].str.lower() == pitch_l]

    if needs_lights is not None:
        subset = subset[subset["Lights"] == ("Yes" if needs_lights else "No")]

    if needs_parking is not None:
        subset = subset[subset["Parking"] == ("Yes" if needs_parking else "No")]

    if max_price is not None:
        subset = subset[subset["Price (LKR)"] <= max_price]

    if subset.empty:
        print("No grounds match these filters.")
        return

    # Preprocess for model
    temp = subset.copy()
    temp = temp.drop(columns=["Ground Name"])

    for col in yes_no_cols:
        temp[col] = temp[col].map({"Yes": 1, "No": 0})

    for col in numeric_cols:
        temp[col] = temp[col].fillna(num_medians[col])

    for col in categorical_cols:
        temp[col] = temp[col].fillna(cat_modes[col])

    temp_encoded = pd.get_dummies(temp, columns=categorical_cols, drop_first=True)
    temp_encoded = temp_encoded.reindex(columns=model_features, fill_value=0)

    temp_scaled = scaler.transform(temp_encoded)

    scores = rf_model.predict_proba(temp_scaled)[:, 1]

    subset["score"] = scores
    subset = subset.sort_values(by="score", ascending=False)

    return subset.head(top_n)
```

```python
recommend_grounds(
    province="Western",
    pitch="matting",
    max_price=80000,
    top_n=5
)
```

```python
recommend_grounds(
    ground_type="Public",
    needs_lights=True,
    needs_parking=True,
    top_n=10
)
```
