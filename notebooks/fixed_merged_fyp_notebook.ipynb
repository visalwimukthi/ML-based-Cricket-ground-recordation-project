```python
def recommend_grounds(user_features, date, team_name=None, city=None, top_n=5):

    data = df.copy()

    # make all columns lowercase 
    data.columns = data.columns.str.lower().str.replace(" ", "_")

    #  FILTER BY CITY 
    if city and "city" in data.columns:
        data = data[data["city"].str.lower() == city.lower()]

    #  TEAM PERFORMANCE 
    if team_name:
        team_data = team_stats[team_stats["team_name"] == team_name].copy()

        if not team_data.empty:
            team_data["win_rate"] = team_data["wins"] / team_data["matches_played"]

            data = data.merge(team_data, on="ground_name", how="left")

            data["win_rate"] = data["win_rate"].fillna(0).astype(float)
            data["matches_played"] = data["matches_played"].fillna(0).astype(float)

            data = data.drop_duplicates(subset=["ground_name"])
        else:
            data["win_rate"] = 0
    else:
        data["win_rate"] = 0

    # WEATHER 
    if "city" in data.columns:
        data["weather"] = data["city"].apply(lambda x: get_weather(x, date))
    else:
        data["weather"] = "unknown"

    # WEATHER SCORE 
    def weather_score(condition):
        condition = str(condition).lower()
        if "rain" in condition or "storm" in condition:
            return 0
        elif "cloud" in condition:
            return 0.7
        else:
            return 1

    data["weather_score"] = data["weather"].apply(weather_score)

    # REMOVE BOOKED GROUNDS 
    if 'bookings' in globals() and not bookings.empty:
        booked = bookings[bookings["date"] == date]["ground_name"]
        data = data[~data["ground_name"].isin(booked)]

    # ML SCORING 
    categorical_cols = ["city", "district", "province", "ground_type", "pitch"]

    # One-hot encoding
    data_encoded = pd.get_dummies(data, columns=categorical_cols)

    # Align with trained model features
    data_encoded = data_encoded.reindex(columns=model_features, fill_value=0)

    # Convert to numeric
    data_encoded = data_encoded.apply(pd.to_numeric, errors='coerce').fillna(0)

    # Scale
    X_scaled = scaler.transform(data_encoded)

    # Predict
    scores = model.predict_proba(X_scaled)[:, 1]

    # Assign ML score
    data["ml_score"] = scores

    # FINAL HYBRID SCORE
    data["final_score"] = (
        data["ml_score"] * 0.4 +
        data["win_rate"] * 0.4 +
        data["weather_score"] * 0.2
    )

    # RETURN TOP RESULTS 
    return data.sort_values(by="final_score", ascending=False).head(top_n)
```

```python
def run_system():
    city = input("Enter city: ").strip()
    date = input("Enter date (YYYY-MM-DD): ").strip()

    team_input = input("Enter team name (optional): ").strip()
    team_name = team_input if team_input != "" else None

    user_features = {
        "price_(lkr)": 120000,
        "rating": 4.5
    }

    # ---------------- RECOMMEND ----------------
    result = recommend_grounds(
        user_features=user_features,
        date=date,
        team_name=team_name,
        city=city,
        top_n=5
    )

    if result.empty:
        print("❌ No grounds found")
        return

    # ---------------- APPLY TEAM-BASED SCORING ----------------
    if team_name is not None:
        result["team_boost"] = result["win_rate"] * 0.2
    else:
        result["team_boost"] = 0

    # 🔥 Updated Final Score (IMPORTANT)
    result["final_score"] = (
        (result["ml_score"] * 0.5) +
        (result["win_rate"] * 0.3) +
        (result["team_boost"] * 0.2)
    )

    # Sort again after applying team influence
    result = result.sort_values(by="final_score", ascending=False).reset_index(drop=True)

    # ---------------- EXPLAIN AI FUNCTION ----------------
    def generate_explanation(row):
        reasons = []

        if row["final_score"] > 0.75:
            reasons.append("high overall score")

        if row["ml_score"] > 0.75:
            reasons.append("strong ML prediction")

        if row["win_rate"] > 0.70:
            reasons.append("good historical win rate")

        if team_name:
            reasons.append("team performance considered")

        if "clear" in str(row["weather"]).lower():
            reasons.append("clear weather conditions")

        if "cloud" in str(row["weather"]).lower():
            reasons.append("acceptable weather conditions")

        if "rain" in str(row["weather"]).lower():
            reasons.append("weather is risky")

        return ", ".join(reasons)

    # ---------------- BEST MATCH ----------------
    top = result.iloc[0]

    print("\n🎯 Ground Recommendation Result\n")

    print("⭐ Best Match:")
    print(f"Ground: {top['ground_name']}")
    print(f"ML Score: {top['ml_score']:.2f}")
    print(f"Win Rate: {top['win_rate']:.2f}")
    print(f"Final Score: {top['final_score']:.2f}")
    print(f"Weather: {top['weather']}")
    print(f"Explanation: {generate_explanation(top)}")

    # ---------------- WEATHER CHECK ----------------
    bad_weather = ["rain", "heavy rain", "storm"]
    alternatives = result.iloc[1:4]

    if any(w in str(top["weather"]).lower() for w in bad_weather):

        print("\n⚠️ Warning:")
        print("This ground is not suitable due to bad weather.\n")

        if not alternatives.empty:
            print("👉 Top 3 Best Alternative Grounds:")
            print("These are the next best options based on score and weather:\n")

            for i, (_, alt) in enumerate(alternatives.iterrows(), 1):
                print(f"\nAlternative {i}:")
                print(f"Ground: {alt['ground_name']}")
                print(f"ML Score: {alt['ml_score']:.2f}")
                print(f"Win Rate: {alt['win_rate']:.2f}")
                print(f"Final Score: {alt['final_score']:.2f}")
                print(f"Weather: {alt['weather']}")
                print(f"Explanation: {generate_explanation(alt)}")

        print("\n✅ Reason:")
        print("Better weather conditions and suitable for playing.")

    else:
        print("\n✅ This ground is suitable for play.")

    # ---------------- BOOKING ----------------
    print("\n🎯 these are the other best suitable grounds,Select a ground to book from the best options:")
    print(f"0 - {top['ground_name']}")

    for i, (_, alt) in enumerate(alternatives.iterrows(), 1):
        print(f"{i} - {alt['ground_name']}")

    choice = input("\nEnter option (0/1/2/3) or 'no': ").strip().lower()

    if choice in ["0", "1", "2", "3"]:


        os.makedirs("artifacts", exist_ok=True)
        file_path = "artifacts/bookings.csv"

        team_name_final = team_name if team_name else "Anonymous"

        # Select ground
        if choice == "0":
            selected = top
        else:
            index = int(choice) - 1
            if index < len(alternatives):
                selected = alternatives.iloc[index]
            else:
                print("❌ Invalid selection")
                return

        # Save booking
        new_df = pd.DataFrame([{
            "ground_name": selected["ground_name"],
            "date": date,
            "booked_by": team_name_final
        }])

        if os.path.exists(file_path):
            new_df.to_csv(file_path, mode='a', header=False, index=False)
        else:
            new_df.to_csv(file_path, index=False)

        print(f"✅ Booking saved for: {selected['ground_name']}")

    elif choice == "no":
        print("❌ Booking cancelled")

    else:
        print("⚠️ Invalid input")
```

```python
def check_availability(ground_name, date):
    df = pd.read_csv(FILE_PATH)

    clash = df[
        (df["ground_name"] == ground_name) &
        (df["date"] == date)
    ]

    return clash.empty
```

```python
def show_availability_for_date(date):
    df_bookings = pd.read_csv(FILE_PATH)

    # 🔥 Load all grounds from your dataset
    grounds_df = pd.read_csv("artifacts/Ground_clean_dataset.csv")

    print(f"\n📅 Availability for {date}:\n")

    for ground in grounds_df["ground_name"].unique():
        available = df_bookings[
            (df_bookings["ground_name"] == ground) &
            (df_bookings["date"] == date)
        ].empty

        status = "🟢 Available" if available else "🔴 Booked"
        print(f"{ground}: {status}")
```

```python
def calendar_view_by_month(year, month):
    df = pd.read_csv(FILE_PATH)

    if df.empty:
        print("📅 No bookings yet")
        return

    # Convert date column to datetime
    df["date"] = pd.to_datetime(df["date"], errors='coerce')

    # Filter by year and month
    df_filtered = df[
        (df["date"].dt.year == year) &
        (df["date"].dt.month == month)
    ]

    if df_filtered.empty:
        print(f"📅 No bookings for {year}-{month:02d}")
        return

    print(f"\n📅 Calendar View for {year}-{month:02d}:\n")

    grouped = df_filtered.groupby(df_filtered["date"].dt.date)

    for date, group in grouped:
        print(f"\n📆 {date}")
        for _, row in group.iterrows():
            print(f"   🏟️ {row['ground_name']} (Booked by {row['booked_by']})")
```

```python
## import uuid
from datetime import datetime


# ---------------- PLAYER ----------------
class Player:
    def __init__(self, name):
        self.name = name
        self.runs = 0
        self.balls = 0
        self.wickets = 0
        self.runs_conceded = 0


# ---------------- TEAM ----------------
class Team:
    def __init__(self, name):
        self.name = name
        self.players = []
        self.score = 0
        self.wickets = 0


# ---------------- SAVE FUNCTIONS ----------------
def save_all(match):
    os.makedirs("artifacts", exist_ok=True)

    # Ball log
    pd.DataFrame(match.ball_log).to_csv(
        "artifacts/match_log.csv", mode='a', index=False, header=False
    )

    # Batting
    bat = []
    for t in [match.teamA, match.teamB]:
        for p in t.players:
            if p.balls > 0:
                bat.append({
                    "match_id": match.match_id,
                    "team": t.name,
                    "player": p.name,
                    "runs": p.runs,
                    "balls": p.balls
                })
    pd.DataFrame(bat).to_csv("artifacts/batting_scorecard.csv", mode='a', index=False, header=False)

    # Bowling
    bowl = []
    for t in [match.teamA, match.teamB]:
        for p in t.players:
            if p.runs_conceded > 0 or p.wickets > 0:
                bowl.append({
                    "match_id": match.match_id,
                    "team": t.name,
                    "player": p.name,
                    "runs_conceded": p.runs_conceded,
                    "wickets": p.wickets
                })
    pd.DataFrame(bowl).to_csv("artifacts/bowling_scorecard.csv", mode='a', index=False, header=False)

    # Summary CSV
    pd.DataFrame([{
        "match_id": match.match_id,
        "date": datetime.now(),
        "teamA": match.teamA.name,
        "teamA_score": f"{match.teamA.score}/{match.teamA.wickets}",
        "teamB": match.teamB.name,
        "teamB_score": f"{match.teamB.score}/{match.teamB.wickets}",
        "winner": match.winner
    }]).to_csv("artifacts/match_summary.csv", mode='a', index=False, header=False)

    print("✅ ALL DATA SAVED")


# ---------------- MATCH SUMMARY DISPLAY ----------------
def show_match_summary(match):

    print("\n" + "="*40)
    print("🏏 MATCH SUMMARY")
    print("="*40)

    print(f"\n{match.teamA.name}: {match.teamA.score}/{match.teamA.wickets}")
    print(f"{match.teamB.name}: {match.teamB.score}/{match.teamB.wickets}")

    print(f"\n🏆 Winner: {match.winner}")

    all_players = match.teamA.players + match.teamB.players

    # Top batsman
    top_batsman = max(all_players, key=lambda x: x.runs)
    print(f"\n🔥 Top Batsman: {top_batsman.name} ({top_batsman.runs} runs)")

    # Top bowler
    top_bowler = max(all_players, key=lambda x: x.wickets)
    print(f"🎯 Top Bowler: {top_bowler.name} ({top_bowler.wickets} wickets)")

    print("="*40)


# ---------------- MATCH ----------------
class Match:
    def __init__(self, teamA, teamB, overs):
        self.teamA = teamA
        self.teamB = teamB
        self.overs = overs
        self.ball_log = []
        self.match_id = str(uuid.uuid4())[:8]

    def get_input(self):
        valid = ["0","1","2","3","4","6","W","WD","NB"]
        while True:
            b = input("Ball (0,1,2,3,4,6,W,WD,NB): ").upper()
            if b in valid:
                return b
            print("❌ Invalid!")

    def play(self, batting, bowling, innings):

        print(f"\n🏏 Innings {innings}: {batting.name}")

        striker = Player(input("Enter Striker name: "))
        non_striker = Player(input("Enter Non-Striker name: "))
        batting.players.extend([striker, non_striker])

        balls = 0
        bowler = None

        while balls < self.overs * 6 and batting.wickets < 10:

            if balls % 6 == 0:
                bowler = Player(input("\nEnter Bowler name: "))
                bowling.players.append(bowler)

            over = balls//6 + 1
            ball_no = balls%6 + 1

            print(f"\nOver {over}.{ball_no}")
            print(f"{batting.score}/{batting.wickets}")

            ball = self.get_input()

            # LOG
            self.ball_log.append({
                "match_id": self.match_id,
                "innings": innings,
                "over": over,
                "ball": ball_no,
                "batsman": striker.name,
                "bowler": bowler.name,
                "event": ball
            })

            if ball == "W":
                batting.wickets += 1
                bowler.wickets += 1
                print(f"{striker.name} OUT!")

                striker = Player(input("Enter next batsman: "))
                batting.players.append(striker)

            elif ball in ["WD","NB"]:
                batting.score += 1
                continue

            else:
                runs = int(ball)
                batting.score += runs
                striker.runs += runs
                striker.balls += 1
                bowler.runs_conceded += runs

                if runs % 2 == 1:
                    striker, non_striker = non_striker, striker

            balls += 1

            if balls % 6 == 0:
                striker, non_striker = non_striker, striker

        return batting.score

    def start(self):
        print(f"\n🆔 Match ID: {self.match_id}")

        target = self.play(self.teamA, self.teamB, 1) + 1
        print(f"\n🎯 Target: {target}")

        score2 = self.play(self.teamB, self.teamA, 2)

        if score2 >= target:
            print(f"\n🎉 {self.teamB.name} WON!")
            self.winner = self.teamB.name
        else:
            print(f"\n🏆 {self.teamA.name} WON!")
            self.winner = self.teamA.name

        # SAVE
        save_all(self)

        # SHOW SUMMARY 🔥
        show_match_summary(self)


# ---------------- RUN ----------------
os.makedirs("artifacts", exist_ok=True)

teamA_name = input("Enter Team A: ")
teamB_name = input("Enter Team B: ")

overs = int(input("Enter overs (2/20/50): "))

teamA = Team(teamA_name)
teamB = Team(teamB_name)

match = Match(teamA, teamB, overs)
match.start()
```

```python

# Load dataset
df = pd.read_csv("../artifacts/Ground_clean_dataset.csv")

# Load trained ML components
rf_model = joblib.load("../artifacts/rf_model.joblib")
scaler = joblib.load("../artifacts/scaler.joblib")
model_features = joblib.load("../artifacts/model_features1.joblib")

print("Dataset and model loaded successfully")
```

```python
# Columns used in training
yes_no_cols = ["Lights", "Parking", "Washrooms", "Pavilion", "Scoreboard"]
numeric_cols = ["Length", "Width", "Rooms", "Parking Cap.", "Price (LKR)", "Rating"]
categorical_cols = ["City", "District", "Province", "Ground Type", "Pitch"]

# Pre-calc medians & modes
num_medians = {col: df[col].median() for col in numeric_cols}
cat_modes = {col: df[col].mode()[0] for col in categorical_cols}

print("Helper stats ready")
```

### normal Ground Search

```python
def search_grounds(
    city=None,
    district=None,
    province=None,
    ground_type=None,
    pitch=None,
    needs_lights=None,
    needs_parking=None,
    max_price=None
):

    results = df.copy()

    # Normalize text inputs to lowercase 
    city_l        = city.lower().strip()        if city else None
    district_l    = district.lower().strip()    if district else None
    province_l    = province.lower().strip()    if province else None
    ground_type_l = ground_type.lower().strip() if ground_type else None
    pitch_l       = pitch.lower().strip()       if pitch else None

    if city_l:
        results = results[results["City"].str.lower() == city_l]

    if district_l:
        results = results[results["District"].str.lower() == district_l]

    if province_l:
        results = results[results["Province"].str.lower() == province_l]

    if ground_type_l:
        results = results[results["Ground Type"].str.lower() == ground_type_l]

    if pitch_l:
        results = results[results["Pitch"].str.lower() == pitch_l]

    if needs_lights is not None:
        results = results[results["Lights"] == ("Yes" if needs_lights else "No")]

    if needs_parking is not None:
        results = results[results["Parking"] == ("Yes" if needs_parking else "No")]

    if max_price is not None:
        results = results[results["Price (LKR)"] <= max_price]

    if results.empty:
        print("No grounds found for these filters.")
        return

    cols_to_show = [
        "Ground Name",
        "City",
        "District",
        "Province",
        "Ground Type",
        "Pitch",
        "Lights",
        "Parking",
        "Price (LKR)",
        "Rooms",
        "Rating"
    ]
    return results[cols_to_show]
```

### ML-based Recommendation

```python
def recommend_grounds(
    city=None,
    district=None,
    province=None,
    ground_type=None,
    pitch=None,
    needs_lights=None,
    needs_parking=None,
    max_price=None,
    top_n=5
):
    subset = df.copy()

    # Normalize text inputs
    city_l        = city.lower().strip()        if city else None
    district_l    = district.lower().strip()    if district else None
    province_l    = province.lower().strip()   if province else None
    ground_type_l = ground_type.lower().strip() if ground_type else None
    pitch_l       = pitch.lower().strip()       if pitch else None

    # Apply filters
    if city_l:
        subset = subset[subset["City"].str.lower() == city_l]

    if district_l:
        subset = subset[subset["District"].str.lower() == district_l]

    if province_l:
        subset = subset[subset["Province"].str.lower() == province_l]

    if ground_type_l:
        subset = subset[subset["Ground Type"].str.lower() == ground_type_l]

    if pitch_l:
        subset = subset[subset["Pitch"].str.lower() == pitch_l]

    if needs_lights is not None:
        subset = subset[subset["Lights"] == ("Yes" if needs_lights else "No")]

    if needs_parking is not None:
        subset = subset[subset["Parking"] == ("Yes" if needs_parking else "No")]

    if max_price is not None:
        subset = subset[subset["Price (LKR)"] <= max_price]

    if subset.empty:
        print("No grounds match these filters.")
        return

    # Preprocess for model
    temp = subset.copy()
    temp = temp.drop(columns=["Ground Name"])

    for col in yes_no_cols:
        temp[col] = temp[col].map({"Yes": 1, "No": 0})

    for col in numeric_cols:
        temp[col] = temp[col].fillna(num_medians[col])

    for col in categorical_cols:
        temp[col] = temp[col].fillna(cat_modes[col])

    temp_encoded = pd.get_dummies(temp, columns=categorical_cols, drop_first=True)
    temp_encoded = temp_encoded.reindex(columns=model_features, fill_value=0)

    temp_scaled = scaler.transform(temp_encoded)

    scores = rf_model.predict_proba(temp_scaled)[:, 1]

    subset["score"] = scores
    subset = subset.sort_values(by="score", ascending=False)

    return subset.head(top_n)
```

```python
recommend_grounds(
    province="Western",
    pitch="matting",
    max_price=80000,
    top_n=5
)
```
